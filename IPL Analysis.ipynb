```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
```

```python
ipl=pd.read_csv("matches.csv")
ipl.head()
```

```python
ipl.shape
```

```python
ipl["player_of_match"].value_counts()
```

```python
ipl["player_of_match"].value_counts()[0:5]
```

```python
list(ipl["player_of_match"].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()),list(ipl["player_of_match"].value_counts()[0:5]),color="brown")
plt.xlabel("Name of Players")
plt.ylabel("Runs")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl["player_of_match"].value_counts()[0:10].keys()),list(ipl["player_of_match"].value_counts()[0:10]),color="g")
plt.xlabel("Name of Players")
plt.ylabel("Runs")
plt.show()
```

```python
batting_first=ipl[ipl["win_by_runs"]!=0]
```

```python
batting_first.head()
```

```python
batting_first.shape
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first["win_by_runs"],color="orange")
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
batting_first["winner"].value_counts()
```

```python
plt.figure(figsize=(5,5))
plt.bar(list(batting_first["winner"].value_counts()[0:3].keys()),list(batting_first["winner"].value_counts()[0:3]),color=["b","brown","orange"])
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(list(batting_first["winner"].value_counts()),labels=list(batting_first["winner"].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(list(batting_first["winner"].value_counts()[0:5]),labels=list(batting_first["winner"].value_counts()[0:5].keys()),autopct="%0.1f%%")
plt.title("Pie-chart representation")
plt.show()
```
