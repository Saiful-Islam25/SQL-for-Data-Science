```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("COVID-19 BD Dataset-10 april.csv")
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
sns.barplot(x="Total confirmed cases",y="Total deaths",data=data,palette='Blues_d')
plt.show()
```

```python
sns.barplot(x="Total deaths",y="Total Tests",data=data,palette='rocket')
plt.show()
```

```python
sns.barplot(x="Total deaths",y="Total confirmed cases",data=data,palette='vlag')
plt.show()
```

```python
sns.scatterplot(x="Total deaths",y="Total confirmed cases",data=data,hue='Total confirmed cases')
plt.show()
```

```python
sns.scatterplot(x="Total confirmed cases",y="Total deaths",data=data,hue='Total deaths')
plt.show()
```

```python
sns.jointplot(x="Total confirmed cases",y="Total deaths",data=data,hue='Total deaths')
plt.show()
```
